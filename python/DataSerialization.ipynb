```python
# getting data from dictionary to string and then from string to dictionary back

import json

data={"last_name":"Sheikh","middle_name":"ahmad","first_name":"fairoz","age":10}
```

```python
string_data=json.dumps(data)
string_data
```

```python
type(string_data)
```

```python
dict_data=json.loads(string_data)
dict_data
```
