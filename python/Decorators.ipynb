```python
# Closure means function inside a fucntion

def message_function():
    name="fairoz"
    def sub_function():
        print(f"Welome to they course {name}")
        print("Study hard")
    return sub_function()
```

```python
message_function()
```

```python
def parent_function(func,listtt):
    def child_function():
        print("I am a closure function")
        print(func(listtt))
    return child_function()

parent_function(len,[1,2,3,5,5,6,])
```

```python
def parent_function(func,listtt):
    def child_function():
        print("I am a closure function")
        print(func(listtt))
    return child_function()

parent_function(max,[1,2,3,5,5,6,10])
```

Decorators are used to add functionality without making changes to the original funcction 

### Decorators

```python
def main_welcome(func):
    def sub_welcome():
        print("Hello studentts welcome to the course")
        print(func())
        print("earn more")
    return sub_welcome()
```
