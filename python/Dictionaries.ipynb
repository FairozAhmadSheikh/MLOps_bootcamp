```python
empty_dict={}
empty_dict2=dict()

print(empty_dict)
print(empty_dict2)
```

```python
student={
    "name":"Feroz",
    "age":20,
    "grade":24.4,
}

print(student)
```

```python
# Accessing dictionary elements

student['grade']
```

```python
student.get('grade')
```

```python
print(student.get('lastname'))   
```

```python
print(student.get('lastname',"Not available"))   
```

```python
# update and add new keys
# update existing
student['age']=28

# add new pair
student['lastname']="Sheikh"
print(student)
```

```python
# delete keys
del student['lastname']
print(student)
```

```python
# Methods of dictionaries
student.keys()
```

```python
student.values()
```

```python
student.items()
```

```python
for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key, value in student.items():
    print(key," == ",value)
```

```python
# Making copies
 
student2=student  # If change made to student2 that will change student also bcoz same memory allocation
```

```python
student2
```

```python
student2["name"]="arshad"
print(student2)
print(student)
```

```python
# Use shallow copy instead 

student3=student.copy()
student3
```

```python
student3['name']='Feroz'
print(student3)
print(student)
```

```python
# nested dictionaries

student_new={
    'feroz':{'rollno':30,"age":20,'lastname':"sheikh"},
    'arshad':{'rollno':3,"age":19,'lastname':"kath"},
}
print(student_new)
```

```python
print(student_new['arshad']["lastname"])
```
