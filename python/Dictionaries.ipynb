```python
# update and add new keys
# update existing
student['age']=28

# add new pair
student['lastname']="Sheikh"
print(student)
```

```python
# Making copies
 
student2=student  # If change made to student2 that will change student also bcoz same memory allocation
```

```python
student2
```

```python
student2["name"]="arshad"
print(student2)
print(student)
```

```python
# Use shallow copy instead 

student3=student.copy()
student3
```

```python
student3['name']='Feroz'
print(student3)
print(student)
```

```python
# nested dictionaries

student_new={
    'feroz':{'rollno':30,"age":20,'lastname':"sheikh"},
    'arshad':{'rollno':3,"age":19,'lastname':"kath"},
}
print(student_new)
```

```python
print(student_new['arshad']["lastname"])
```

```python
for student_name,student_details in student_new.items():
    for keys , values in student_details.items():
        print(f'The name of student is {student_name} and details are {keys} = {values}')
```

Example to calculate frequency of numbers

```python
numbers=[1,2,3,3,3,4,5,6,6,6,6,7,7,8,9,10]
frequency={}

for item in numbers:
    if item in frequency:
        frequency[item]+=1
    else:
        frequency[item]=1

print(frequency)
```

```python
# Merge two dictionaries
dict1={"a":1,"b":2,"c":3}
dict2={"a":2,"d":2,"c":5}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
