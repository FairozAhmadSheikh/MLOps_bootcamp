Access Modifiers

    * public
    * private
    * protected

```python
# Public variables 
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

p1=Person("feroz",21)
print(p1.name)
print(p1.age)
```

```python
# Public and private
class Person:
    def __init__(self,name,salary):
        self.name=name  # public
        self.__salary=salary # protected means cant be called from outside class , need to define getter and setter methods

p2=Person("sheikh",350000)
print(p1.name) # accessable and not the salary 
```

```python
# Protected   means dervived class can access those variables
class Person:
    def __init__(self,name,age,gender):
        self._name=name
        self._age=age
        self._gender=gender

class Employee(Person):
     def __init__(self,name,age,gender):
        super().__init__(name,age,gender)


```

```python
emp=Employee("krishi darshan",22,'Male')
```

Encapsulation with getter and setter methods

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name
        self.__age=age
        self.__gender=gender
    # Getter for name
    def get_name(self):
        print(f'Hello {self.__name}')
    # getter for age
    def get_age(self):
        print(f'{self.__name} your age  is {self.__age}')
    # setter for name
    def set_name(self,new_name):
        self.__name=new_name
    # setter for age
    def set_age(self,new_age):
        if new_age>0
            self.__age=new_age
        else:
            print("Please eneter a valid age")
    
```
