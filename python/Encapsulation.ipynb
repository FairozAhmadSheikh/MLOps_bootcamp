Access Modifiers

    * public
    * private
    * protected

```python
# Public variables 
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

p1=Person("feroz",21)
print(p1.name)
print(p1.age)
```

```python
# Public and private
class Person:
    def __init__(self,name,salary):
        self.name=name  # public
        self.__salary=salary # protected

p2=Person("sheikh",350000)
print(p1.name) # accessable and not the salary 
```
