```python
# ZeroDivisonError
try:
    a=10
    b=0
    a/b
except ZeroDivisionError as ex :
    print(ex)
    print("You cant divide a number with zero")
```

```python
# handling unknown errors 
try:
    a=10
    b=20

    a/b

    c="this"
    a/c
except ZeroDivisionError :
    print("Cant divide with zero")
except Exception as ex:
    print(ex)
```

```python
# Try else except and finally 

try:
    num=int(input("Enter the number : "))
    result= 10/num
except ValueError :
    print("That is not a valid number")
else:
    print(f'This is the {result} expression calucalted  this block runs if no errors i try block')
finally:
    print(f'This block will execute even in case of errors also')
```
