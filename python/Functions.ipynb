```python
# Syntax of functions

def even_odd_check(number):
    """
    This Fucntion checks for numbers in even odd,
    Number : Provide a Number here
    """
    if number%2 == 0:
        print("Even")
    else:
        print("Odd")


even_odd_check(45)
even_odd_check(44)
even_odd_check(2)
```

```python
# Variable length parameter

def display_numbers(*args):
    for  item in args :
        print(f" {item} and is displayed sucessfully")
```

```python
display_numbers(["feroz","sheikh",'arshad','apple','mango'])
```

# Examples
* Keyword and Positional 

```python
# keyword and positional arguments both
def display_details(*args,**kwargs):
    """
    This functions accepts a list or items 
    Paramters:
    *args=takes arguments list 
    **kwargs= takes dictionaries or key value pairs
    """
    for item in args:
        print(item)
    for key , values in kwargs.items():
        print(f'{key} : {values}')
```

```python
listted=["feroz","sheikh",'arshad','apple','mango']
student={
        'name':"ali",
        'age':21,
        'roll':1
        }
display_details(listted,student)
```

Example Total cost of items

```python
cart=[
    {"name":'apple','price':0.3,'qunatity':10},
    {"name":'mango','price':0.5,'qunatity':6},
    {"name":'banana','price':0.8,'qunatity':5},
]

def generate_bill(cart):
    total_billl=0
    for items in cart :
        rate=items['price']*items['qunatity']
        total_billl+=rate
    print(f'total bill is {total_billl} $ ')
```

```python
generate_bill(cart)
```
