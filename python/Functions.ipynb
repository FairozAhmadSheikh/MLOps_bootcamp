```python
# Syntax of functions

def even_odd_check(number):
    """
    This Fucntion checks for numbers in even odd,
    Number : Provide a Number here
    """
    if number%2 == 0:
        print("Even")
    else:
        print("Odd")


even_odd_check(45)
even_odd_check(44)
even_odd_check(2)
```

```python
def greet(name):
    """
    This function accepts a name and greets 
    """
    return f'Hello and Welcome {name}'

print(greet("feroz"))
```

```python
# Variable length parameter

def display_numbers(*args):
    for  item in args :
        print(f" {item} and is displayed sucessfully")
```

```python
display_numbers(["feroz","sheikh",'arshad','apple','mango'])
```
