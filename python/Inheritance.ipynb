```python
# Lets see inheritance 
class Car:
    def __init__(self,doors,engine_type):
        self.doors=doors
        self.engine_type=engine_type
    
    def drive(self):
        print(f'The car has been started , {self.engine_type}')

class Tesla(Car):
      def __init__(self,doors,engine_type,self_driving):
           super().__init__(doors,engine_type)
           self.self_driving=self_driving
```

```python
tesla=Tesla(4,"electric",True)

tesla.drive()
```
