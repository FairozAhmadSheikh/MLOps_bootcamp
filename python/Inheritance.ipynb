```python
# Lets see inheritance 
class Car:
    def __init__(self,doors,engine_type):
        self.doors=doors
        self.engine_type=engine_type
    
    def drive(self):
        print(f'The car has been started , {self.engine_type}')

class Tesla(Car):
      def __init__(self,doors,engine_type,self_driving):
           super().__init__(doors,engine_type)
           self.self_driving=self_driving
```

```python
tesla=Tesla(4,"electric",True)

tesla.drive()
```

# Multiple inheritance

```python
class Animal:
     def __init__(self,name):
          self.name=name
     
     def speak(self):
          print("Subclass will implement this")
```

```python
class Pet:
      def __init__(self,owner):
            self.owner=owner
    
        
```

```python
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        print(f'Woof Woof!')
```

```python
d1=Dog("Labrador","sameer")
d1.speak()
```
