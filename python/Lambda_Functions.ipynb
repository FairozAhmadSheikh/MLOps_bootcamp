```python
# Syntax for lambda functions

# lambda arguments : expression
```

```python
addition= lambda x,y:x+y

print(addition(10,12))
```

```python
# even 
even= lambda x:x % 2==0
even(2)
```

# Map Functions

# synatax
map(function, iterateables)

```python
print(list(map(lambda x : x**2,[1,2,3,4,5,6])))
```

```python
def squares(num):
    return num**2
```

```python
print(list(map(squares,[1,2,3,4,5,6])))
```

```python
# map multiple iterateables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
added_numbers
```

```python
chars=['1','2','3','4']
nums =list(map(int,chars))
nums
```
