Numpy Vectorized Opearations


```python
arr1=np.array([1,2,3,4,5,6])
arr2=np.array([1,2,3,4,5,6])

# Element wise operation

print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```
