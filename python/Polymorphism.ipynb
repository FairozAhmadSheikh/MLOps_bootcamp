```python
# Method Overriding

class Animal:
    def speak(self):
        print("This should have sound of animal")

class Dog(Animal):
     def speak(self):
         print("woof woof")

class Cat(Animal):
    def speak(self):
        print("meow mewo ")

d1=Dog()
c1=Cat()
d1.speak()
c1.speak()
d1.speak()
```
