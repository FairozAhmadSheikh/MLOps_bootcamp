```python
# Operations on sets 

set1={1,2,3,4,5}
set2={3,4,6,7,8}

set1.add(6)
print(set1)

set1.remove(6)
print(set1)

set1.discard(0)  # used to delete an element even if an element does not exist it does not show errors

set1.pop()
print(set1)

set1.clear()
print(set1)
```
