```python
from abc import ABC,abstractmethod

class Vehicle:
    @abstractmethod
    def start_engine(self):
        pass
```

```python
class Car(Vehicle):
    def start_engine(self):
        return "car engine started"

class Bike(Vehicle):
    def start_engine(self):
        return "Bike Engine started"
```

```python
car=Car()
car.start_engine()
```

```python
bike=Bike()
bike.start_engine()
```
