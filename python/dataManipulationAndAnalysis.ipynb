```python
df2=pd.read_csv('food.csv')
df2.isnull().any()
```

```python
# filled dataset
df_filled=df2.fillna(0)
df_filled.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df2['Yoghurt']=df2['Yoghurt'].fillna(df2['Yoghurt'].mean())
df2.isnull().sum()
```

```python
df2['Biscuits']=df2['Biscuits'].fillna(df2['Biscuits'].mean())
df2['Sweetener']=df2['Sweetener'].fillna(df2['Sweetener'].median())
```

```python
df2.isnull().sum()
```
