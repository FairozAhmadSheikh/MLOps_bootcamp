```python
# append mode

with open('example.txt','a') as file:
    file.write("\n Previous text stays with this also included")
```

```python
# adding multiple lines now

# using w+ mode which means file will be created if the file does not exist 

lines=['\n this line will be added \n','This also will be added \n', 'This line also will be added \n ']

with open('example.txt','w+') as file:
     file.writelines(lines)
     print("sucessfully wrote lines to file ")
```

```python
# Read and write at same time using seek 

with open("example.txt",'w+') as file:
    file.write("This is a text new as fresh \n")
    file.write("This is a  new  line added  \n")

    # set pointer back to first place 
    file.seek(0)
    content=file.read()
    print(content)
```
