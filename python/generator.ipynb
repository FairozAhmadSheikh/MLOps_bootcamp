```python
def square(n):
    for i in range(n):
        return i ** 2

square(3)
```

```python
def square(n):
    for i in range(n+1):
        yield i **2

for i in square(3):
    print(i)
```

# Reading large file using genrators

```python
def read_large_files():
    with open ('example.txt') as file:
        for line in file:
            yield line
```

```python
# Read all file at once
for line in read_large_files():
    print(line)
```
