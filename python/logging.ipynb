```python
# way 2 
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s %(name)s %(levelname)s  %(message)s',
    datefmt='%Y -%m -%d  %H:%M:%S',
    handlers=[
        logging.FileHandler("applog.log"),
        logging.StreamHandler()
        ]
)

logger=logging.getLogger("Arithamtic app")
```

```python
def add(a,b):
    result=a+b
    logger.debug(f'adding {a} + {b} : {result}')
    return result
```

```python
def sub(a,b):
    result=a-b
    logger.debug(f'subtracting {a} - {b}: {result}')
    return result
```

```python
def div(a,b):
    try:
        result=a/b
        logger.debug(f'dividing {a} / {b} : {result}')
        return result
    except ZeroDivisionError:
        logger.error(f'Zero Division error ')
        return None
    
```

```python
add(10,15)
```

```python
sub(10,15)
```

```python
div(10,0)
```

```python
div(10,10)
```
