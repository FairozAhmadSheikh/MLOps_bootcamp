# Doing a real world example now 
lets make bank system using oops

```python
class BankAccount:
    def __init__(self,name,balance=0):
        self.name = name
        self.balance = balance
    def deposit(self,amount):
        self.amount=amount
        if amount >=0: 
            self.balance+= amount
            print(f"{self.amount} deposited successfully \n Total balance now : {self.balance}")
        else:
            print(f"{self.amount} cannot be added , please enter a valid amount ")
    def withdrawl(self,amount):
        self.amount=amount
        if amount<=self.balance:
            self.balance-=amount
            print(f'{self.amount} withdrawn successfully, Remaining balance is {self.balance}')
        else:
            print('not sufficent funds')
    def get_balance(self):
        print(f'total balcnce {self.balance}  $ ') 
    
```

```python
b1=BankAccount("ali",100)
```

```python
b1.get_balance()
```

```python
b1.deposit(1000)
b1.get_balance()
```

```python
b1.withdrawl(1000)
```

```python
b1.get_balance()
```
